## Ecxploratory Data Analysis

## Updating & Rearranging Columns

```python
# Check the actual column names in your DataFrame
Unemployment.columns
```

```python
Unemployment.columns = Unemployment.columns.str.strip()
Unemployment.columns
```

```python
# Rename columns for easier reference
Unemployment.rename(columns={
    'Estimated Unemployment Rate (%)': 'Unemployment_Rate',
    'Estimated Employed': 'Employed',
    'Estimated Labour Participation Rate (%)': 'Labour_Participation_Rate'}, inplace = True)
```

```python
# Display the first few rows after preprocessing
Unemployment.head()
```

```python
# Visualize unemployment rate over time
plt.figure(figsize=(10, 4))

sns.lineplot(data=Unemployment, x='Date', y='Unemployment_Rate') 

plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
Unemployment ['Unemployment_Rate']
```

```python
plt.figure(figsize=(14, 5))
plt.bar(Unemployment['Region'], Unemployment['Unemployment_Rate'], color='teal')
plt.title('Unemployment Rate by Region', fontsize=16)
plt.xlabel('Region', fontsize=14)
plt.ylabel('Unemployment Rate (%)', fontsize=14)
plt.xticks(rotation=75, fontsize=12)
plt.show()
```

## Area wise Unemployment_Rate

```python
Unemployment ['Area'].value_counts()
```

```python
# area wise unemployment rate using pie chart
plt.figure(figsize=(10,4))
Unemployment.groupby('Area')['Unemployment_Rate'].mean().plot(kind='pie', autopct='%1.1f%%')
plt.title('Area Wise Unemployment_Rate')
plt.show()
```

Observation
- Its clearly visible that Urban Area has more (56.0%) Unemployment_Rate than Rural Area (44.0%)

## Min, Max, Avg Unemployment Rate

```python
avg_unemployment_rate = Unemployment.groupby('Region')['Unemployment_Rate'].mean()
state_with_highest_unemployment = avg_unemployment_rate.idxmax()
high_unemployment_rate = avg_unemployment_rate.max()
state_with_lowest_unemployment = avg_unemployment_rate.idxmin()
low_unemployment_rate = avg_unemployment_rate.min()
print(f"State with Highest Unemployment : {state_with_highest_unemployment}")
print(f"Highest Unemployment Rate : {high_unemployment_rate}")
print(f"State with Lowest Unemployment : {state_with_lowest_unemployment}")
print(f"Low Employment Rate : {low_unemployment_rate}")
```

```python
avg_unemployment_rate
```

## Average Unemployment Rate by Region

```python
# Create the bar plot
plt.figure(figsize=(10, 6))
plt.plot(avg_unemployment_rate.index, avg_unemployment_rate.values, marker='o', label='Unemployment Rate', color='black')
plt.bar(avg_unemployment_rate.index, avg_unemployment_rate.values, label='Unemployment Rate', color='darkgrey')

# Highlight the highest and lowest unemployment rate regions
plt.bar(state_with_highest_unemployment, high_unemployment_rate, color='red', label=f'Highest: {state_with_highest_unemployment}')
plt.bar(state_with_lowest_unemployment, low_unemployment_rate, color='green', label=f'Lowest: {state_with_lowest_unemployment}')

# Add labels and title
plt.title('Average Unemployment Rate by Region', fontsize=16)
plt.xlabel('Region', fontsize=14)
plt.ylabel('Unemployment Rate (%)', fontsize=14)
plt.xticks(rotation=75, fontsize=12)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

Observation 
- Red bar indicates Highest Avg Employment Rate in "Tripura"
- Green bar indicates Lowest Avg Employment Rate in "Meghalaya"
- Grey/black plots indicate the unemployment by region

## Labour_Participation_Rate by Region

```python
plt.figure(figsize=(10, 4))
sns.boxplot(x='Region', y='Labour_Participation_Rate', data = Unemployment ,palette = 'Set1')
plt.title('Labour_Participation_ Rate')
plt.xlabel('Region')
plt.ylabel('Labour_Participation_ Rate')
plt.xticks(rotation = 75)
plt.show()
```

Observation
- This visualization shows the Labour Participation Rate by region
- Through this we have seen the Lowest & Highest Labour Participation Rate by region
